```python
dna_record=SeqIO.read("/content/Covid_sequence.fasta","fasta")
print(dna_record)
```

```python
dna_seq = dna_record.seq
print(dna_seq)
```

```python
len(dna_seq)
```

```python
mrna =dna_seq.transcribe()
```

```python
print(dna_seq[1:10])
print(mrna[1:10])
```

```python
protien = mrna.translate()
print(protien)
```

```python
amino_acids = protien.split("*")
print(amino_acids)
```

```python
ncov_aa = [str(i) for i in amino_acids]
print(ncov_aa)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'amino_acids' :ncov_aa})
df.head()
```

```python
df['count'] = df['amino_acids'].apply(len)
print(df)
```
